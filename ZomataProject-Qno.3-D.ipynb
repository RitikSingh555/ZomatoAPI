```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==216]
#print(df.head())
df["Cuisines"].replace("",np.nan, inplace = True)
df.dropna(inplace=True, subset=["Cuisines"])
```

```python
## Top 10 on the basis of served in max no. of restaurant
mydict={}
for i in df.Cuisines:
    st=i.split(',')
    for i in st:
        mydict[i.strip()]=mydict.get(i.strip(),0)+1
#print(mydict)
k=Counter(mydict)
h=k.most_common(10)
l1=[]
l2=[]
print("Top 10 Cuisines present in the restaurant in the U.S.A")
for i,j in h:
    print(i,":",j)
    l1.append(i)
    l2.append(j)
    
explode=[1.5,1.5,1.5,1.5,1.5,0,0,0,0,0]
plt.title("Top 10 Cuisines In USA")
plt.pie(l2,labels=l1,autopct="%.2f%%",startangle=130,counterclock=False)
plt.show()
    
```
