```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==1]
mydict={}
Cities=set(df.City)
    
```

```python
for i in Cities:
    mydict[i]=len(df[df['City']==i])
#print(mydict)
```

```python
mydict_1={}
df['Vote_Rating']=df['Votes']*df['Aggregate rating']
for i in Cities:
    df1=df[df.City==i]
    mydict_1[i]=sum(df1.Vote_Rating)
```

```python
mydict_2={}
for i in Cities:
    df2=df[df.City==i]
    mydict_2[i]=mydict_1[i]/sum(df2.Votes)
#print(mydict_2)
```

```python
City_Name=[]
No_of_Restaurnats=[]
Weighted_Restaurant_Rating=[]
for i in Cities:
    City_Name.append(i)
    No_of_Restaurnats.append(mydict[i])
    

for i in Cities:
    Weighted_Restaurant_Rating.append(mydict_2[i])
```

```python
colors=np.random.rand(len(City_Name))
plt.xlabel("City Name")
plt.ylabel("Number of Restaurants")
plt.scatter(City_Name,No_of_Restaurnats,No_of_Restaurnats,c=colors,alpha=0.7,edgecolor="black")
plt.title("Bubble Chart")
plt.xticks(rotation = 90)
plt.show()
```
