```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==1]
```

```python
df.City.replace("New Delhi","Delhi-NCR",inplace=True)
df.City.replace("Ghaziabad","Delhi-NCR",inplace=True)
df.City.replace("Noida","Delhi-NCR",inplace=True)
df.City.replace("Gurgaon","Delhi-NCR",inplace=True)
df.City.replace("Faridabad",'Delhi-NCR',inplace=True)
```

```python
df["Cuisines"].replace("",np.nan, inplace = True)
df.dropna(inplace=True, subset=["Cuisines"])
```

```python
df1=df[df.City=='Delhi-NCR']
df2=df[df.City!='Delhi-NCR']
```

```python
mydict_2={}
mydict_3={}
for i in df1.Cuisines:
    st=i.split(',')
    for j in st:
        mydict_2[j.strip()]=mydict_2.get(j.strip(),0)+1
k=Counter(mydict_2)
h=k.most_common(10)
l1=[]
l2=[]
for i,j in h:
    l1.append(i)
    l2.append(j)
```

```python
plt.bar(l1,l2,width=0.5)
plt.xlabel("Top 10 Cuisines in Delhi NCR")
plt.xticks(rotation=90)
plt.show()
```

```python
for i in df2.Cuisines:
    st=i.split(',')
    for j in st:
        mydict_3[j.strip()]=mydict_3.get(j.strip(),0)+1
k1=Counter(mydict_3)
h1=k1.most_common(10)
l3=[]
l4=[]
for i,j in h1:
    l3.append(i)
    l4.append(j)
```

```python
plt.bar(l3,l4,width=0.5)
plt.xlabel("Top 10 Cuisines in Rest of India")
plt.xticks(rotation=90)
plt.show()
```
