```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==1]
df["Cuisines"].replace("",np.nan, inplace = True)
df.dropna(inplace=True, subset=["Cuisines"])
```

## Part-1

```python

np_rating=np.array(df['Aggregate rating'],dtype=float)
np_vote=np.array(df['Votes'],dtype=int)
xt=[1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0]
plt.bar(np_rating,np_vote,width=0.07)
plt.xlabel("Rating")
plt.ylabel("Votes")
plt.title("Bar Graph")
plt.xticks(xt)
plt.show()
```

## Part-2


```python
l=[]
df.Cuisines.replace('',np.nan,inplace=True)
df.dropna(inplace=True,subset=['Cuisines'])

for i in df.Cuisines:
    st=i.split(',')
    l.append(len(st))
```

```python
plt.bar(l,np_rating,width=0.5)
plt.title("Bar Graph")
plt.xlabel("Cuisines")
plt.ylabel("Rating")
plt.show()
```

## Part-3

```python
avgCost=np.array(df["Average Cost for two"],dtype=int)
plt.bar(avgCost,np_rating,width=1)
plt.title("Bar Graph")
plt.xlabel("Average Cost")
plt.ylabel("Rating")
plt.show()
```

## Part-4

```python
mydict_1={}
for i in df.Cuisines:
    st=i.split(',')
    for j in st:
        mydict_1[j.strip()]=mydict_1.get(j.strip(),0)+1
        Avg_rating=[]
no_times_served=mydict_1.values()
for i in mydict_1.keys():
    df1=df[df['Cuisines']==i]
    Avg_rating.append(sum(df1['Aggregate rating'])/mydict_1[i])
plt.bar(no_times_served,Avg_rating,width=10)
plt.xlabel("NO. of times Cuisines Served")
plt.ylabel("Average Rating")
plt.title("Bar Chart")
#plt.xticks(rotation=90)
plt.show()
```
