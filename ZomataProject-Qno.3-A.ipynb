```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==1]
mydict={}
for i in df['Restaurant Name']:
    mydict[i]=mydict.get(i,0)+1
#print(mydict)
k=Counter(mydict)
h=k.most_common(15)
l=[]
l1=[]
print("Top 15 restaurants having maximun no. of Outlets")
for i,j in h:
    print(i,j)
    l.append(i)
    l1.append(j)
```

```python
plt.bar(l,l1,width=0.6)
plt.xticks(rotation=90)
plt.xlabel("Top 15 Restaurnats")
plt.ylabel("No. of outlets")
plt.show()
```
