```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==1]
mydict={}
np_Votes=np.array(df.Votes,dtype=int)
np_Name=np.array(df['Restaurant Name'],dtype=str)
for i in range(len(np_Name)):
    st=np_Name[i]
    vl=np_Votes[i]
    mydict[st]=mydict.get(st,0)+vl
k=Counter(mydict)
h=k.most_common(10)
l=[]
l1=[]
for i,j in h:
    print(i,":",j)
    l.append(i)
    l1.append(j)

```

```python
plt.bar(l,l1,width=0.9)
plt.xticks(rotation="vertical")
plt.xlabel("Top 10 Restaurants")
plt.ylabel("Number of Votes")
plt.title("Bar Graph")
plt.show()
```
