```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

df=pd.read_csv("D:\DRIVE PART 2\coding ninjaa datasets\zomato_file.csv",encoding = "ISO-8859-1")
df=df[df['Country Code']==1]
df.Votes.replace(0,np.nan,inplace=True)
df.dropna(inplace=True, subset=["Votes"])
df['Vote_Rating']=df['Votes']*df['Aggregate rating']
```

```python
locality=set(df.Locality)
mydict_1={}
for i in locality:
    df1=df[df['Locality']==i]
    mydict_1[i]=sum(df1['Vote_Rating'])



mydict_2={}
for i in locality:
    df2=df[df.Locality==i]
    mydict_2[i]=mydict_1[i]/sum(df2.Votes)
#print(mydict_2)
```

```python
l=[]
l1=[]
k=Counter(mydict_2)
h=k.most_common(10)
for i,j in h:
    print(i,":",j)
    l.append(i)
    l1.append(j)
```

```python
plt.bar(l,l1,width=0.6)
plt.xlabel("Bar Graph")
plt.xlabel("Top 10 Locality")
plt.ylabel("Weighted Restuarnt Rating")
plt.xticks(rotation=90)
plt.show()
```
